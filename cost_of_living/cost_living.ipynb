```python
# Calcul de l'indice du coûte de la vie moyen par pays
cost_of_living_index['average_cost_of_living'] = (cost_of_living_index['Cost of Living Index'] + 
            cost_of_living_index['Rent Index'] +
            cost_of_living_index['Groceries Index'] +
            cost_of_living_index['Restaurant Price Index']) / 4

# Afficher les premier résultats
print(cost_of_living_index[['Country', 'average_cost_of_living']])
```

```python
world_energy_df['total_energy_consumption'] = (
    world_energy_df['coal_consumption'] +
    world_energy_df['biofuel_consumption'] +
    world_energy_df['oil_consumption']  # ajouter d'autres sources d'énergie si nécessaire
)
world_energy_df['energy_consumption_per_capita'] = world_energy_df['total_energy_consumption'] / world_energy_df['population']
```

```python
# Calcule du cout moyen du repas
new_cost_of_living['average_meal_cost'] = (
    new_cost_of_living['meal_inexpensive_restaurant'] + 
    new_cost_of_living['meal_for_2people'] + 
    new_cost_of_living['mcmeal_macdonald']
) / 3

print(new_cost_of_living[['country', 'average_meal_cost']])
```

```python
# Indice du prix des produits alimentaires
new_cost_of_living['food_price_index'] = (
    new_cost_of_living['white_bread'] +
    new_cost_of_living['milk_litter'] +
    new_cost_of_living['rice'] +
    new_cost_of_living['eggs'] +
    new_cost_of_living['chiken_fillet'] +
    new_cost_of_living['beef_round'] +
    new_cost_of_living['apples'] +
    new_cost_of_living['banana'] +
    new_cost_of_living['oranges'] +
    new_cost_of_living['tomato'] +
    new_cost_of_living['potato'] +
    new_cost_of_living['onion']
) / 12

print(new_cost_of_living[['country', 'food_price_index']].head())
```

```python
# Calcule du pourcentage du revenu dépensé en alimentation
new_cost_of_living['food_expenditure_percentage'] = (
    new_cost_of_living['food_price_index'] * 30) / new_cost_of_living['average_monthly_net_salary_after_tax']

print(new_cost_of_living[['country', 'food_expenditure_percentage']].head())
```

```python
# Détermination de la disponibilité des produits aliumentaires
# Un seul produits
new_cost_of_living['availability_index'] = (
    new_cost_of_living['rice'] / new_cost_of_living['white_bread']
)

print(new_cost_of_living[['country', 'availability_index']].head())
```

```python
# Indice de santé alimentaire
## Evalue la qualité nutritionnelle des aliment disoinibles en fonction de leur coût, leur disponibilité, et leur impact dur la santé (sucre, sels gras)

new_cost_of_living[('health_food_index')] = (
    new_cost_of_living['mcmeal_macdonald'] / new_cost_of_living['apples']
)

print(new_cost_of_living[['country', 'health_food_index']].head)
```

```python
### Indice de la disparité alimentaire
## Mesure les ecarts entre les prix des produits alimentaires essentiels dans différente région du pays 
disparity = new_cost_of_living.groupby('country')['food_price_index'].std()

print(disparity)
```
